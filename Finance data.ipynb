```python
start = datetime.datetime(2014, 1, 1)
end = datetime.datetime(2024, 1, 1)
```

```python
# Bank of America
BAC = yf.download("BAC", start=start, end=end)

# CitiGroup
C = yf.download("C", start=start, end=end)

# Goldman Sachs
GS = yf.download("GS", start=start, end=end)

# JPMorgan Chase
JPM = yf.download("JPM", start=start, end=end)

# Morgan Stanley
MS = yf.download("MS", start=start, end=end)

# Wells Fargo
WFC = yf.download("WFC", start=start, end=end)
```

```python
BAC.head()
```

```python
GS.head()
```

** Create a list of the ticker symbols (as strings) in alphabetical order. Call this list: tickers**

```python
tickers = ['BAC', 'C', 'GS', 'JPM', 'MS', 'WFC']
```

```python
df = yf.download(tickers, start=start, end=end, group_by="ticker")
```

** Use pd.concat to concatenate the bank dataframes together to a single data frame called bank_stocks. Set the keys argument equal to the tickers list. Also pay attention to what axis you concatenate on.**

```python
bank_stocks = pd.concat([BAC, C, GS, JPM, MS, WFC],axis=1,keys=tickers)
```

** Set the column name levels (this is filled out for you):**

```python
bank_stocks.columns = bank_stocks.columns.droplevel(-1)
```

```python
bank_stocks.columns.names = ['Bank Ticker', 'Stock Info']
```

** Check the head of the bank_stocks dataframe.**

```python
bank_stocks.head()
```

** Create a new empty DataFrame called returns. This dataframe will contain the returns for each bank's stock. returns are typically defined by:**

$$r_t = \frac{p_t - p_{t-1}}{p_{t-1}} = \frac{p_t}{p_{t-1}} - 1$$

```python
returns = pd.DataFrame()
```

** We can use pandas pct_change() method on the Close column to create a column representing this return value. Create a for loop that goes and for each Bank Stock Ticker creates this returns column and set's it as a column in the returns DataFrame.**

```python
for tick in tickers:
    returns[tick+' Return'] = bank_stocks[tick]['Close'].pct_change()
returns.head()
```

## Moving Averages

Let's analyze the moving averages for these stocks in the year 2020. 

** Plot the rolling 30 day average against the Close Price for Bank Of America's stock for the year 2020**

```python
plt.figure(figsize=(12,6))
ax = plt.gca()
# Plot BAC Close Price for the year 2020 on the same axis (ax)
BAC['Close'].loc['2020-01-01':'2020-12-31'].plot(label='BAC CLOSE', color='blue', ax=ax)

# Plot the 30-day rolling average of BAC Close Price on the same axis (ax)
BAC['Close'].loc['2020-01-01':'2020-12-31'].rolling(window=30).mean().plot(label='30 Day Avg', color='orange', ax=ax)
ax.set_title('Bank of America Close Price and 30-Day Rolling Average (2020)')
ax.set_xlabel('Date')
ax.set_ylabel('Price')
ax.legend()
plt.show()
```

** Create a heatmap of the correlation between the stocks Close Price.**

```python
sns.heatmap(bank_stocks.xs(key='Close',axis=1,level='Stock Info').corr(),annot=True)
```

** Use seaborn's clustermap to cluster the correlations together:**
